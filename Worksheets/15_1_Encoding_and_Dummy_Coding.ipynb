## Encoding Data  
When we encode data, we remodel a numeric column in a dataframe to be on a standard scale (0 or 1).   
For example if we had a column of BMI scores, we could encode that column so that all scores greater than or equal to 25 were recoded to the value 1 (bad) and all scores less than 25 were recoded to 0 (good).  

For example:  

` def encode_bmi(df):`       
> `if df['bmi'] >= 25:`  
> >  `return 1`   

>`else`:  
> >`return 0`  

`df["bmi"] = df.apply(encode_bmi, axis=1)`

````python
import pandas as pd
url = "https://github.com/lilaceri/Working-with-data-/blob/342abab10d93c4bf23b5c55a50f189f12a137c5f/Data%20Sets%20for%20code%20divisio/Covid%20Vaccination%20Data.xlsx?raw=true"

# Task 1 
vax = pd.read_excel(url, sheet_name = 'by_country')

# Task 2

def checknacolumns(df):
  nalist = []
  for column in df.columns:
    if df[column].isnull().values.any(): 
      nalist.append(column)
  return nalist

checknacolumns(vax)

# Task 3
def droprows(df, column):
  df = df.dropna(subset = [column])
  return df

vax = droprows(vax, 'total_vaccinations')

# Task 4
vax = droprows(vax, 'daily_vaccinations_per_million')

# Task 5
def getmedian(df, column):
  df = df.sort_values(column)
  middle = len(df) / 2
  middle = int(middle)
  med = df[column].iloc[middle]
  return med

median = getmedian(vax, 'daily_vaccinations_per_million')
print(median)
````

````python
# Task
def encode_daily(df, column, med):
  #median = getmedian(vax, 'daily_vaccinations_per_million')
  
  if df[column] >= med:
    return 1
  else:
    return 0
med = vax['daily_vaccinations_per_million'].median()
print(vax['daily_vaccinations_per_million'].median())
vax['daily_vaccinations_per_million'] = vax.apply(encode_daily, args = ('daily_vaccinations_per_million', med,), axis = 1)
display(vax)
vax.describe()
````
